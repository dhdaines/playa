# Using PLAYA-PDF to extract text with formatting

Because PLAYA-PDF gives you access to text objects, you can do slightly more interesting things than just dumping out an undifferentiated stream of plain text.

For simplicity here we will use a tagged PDF, where the text objects contain explicit whitespace

```python
import playa
from IPython.display import HTML
```

```python
DOCUMENT = "../samples/pdf_structure.pdf"
pdf = playa.open(DOCUMENT)
page = pdf.pages[0]
```

```python
html = []
for text in page.texts:
    font, _, style = text.fontbase.partition("-")
    face = "serif" if "serif" in font.lower() else "sans"
    css = [
        f"font-size: {text.size}px",
        f"font-family: {face}",
    ]
    if "bold" in style.lower():
        css.append("font-weight: bold")
    html.append(f"<p style='{'; '.join(css)}'>{text.chars}</p>")
HTML(" ".join(html))
```
